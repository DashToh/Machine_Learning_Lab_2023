## Lab 1 - Python Basics

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np

# load the famous iris data
irisRaw = load_iris()
 
#Convert data to Panda package
# read iris.data into a pandas DataFrame (df), including column names
iris = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target']],
                     columns= irisRaw['feature_names'] + ['target'])
```

```python
# View the data
iris
```

```python
# Display the number of rows and columns
print(iris.shape)
```

```python
# View basic information for individual columns
iris.describe()
```

```python
sepal_length = iris.groupby('sepal length (cm)')
sepal_length.describe()
```

```python
sepal_width = iris.groupby('sepal width (cm)')
sepal_width.describe()
```

```python
petal_length = iris.groupby('petal length (cm)')
petal_length.describe()
```

```python
petal_width = iris.groupby('petal width (cm)')
petal_width.describe()
```

```python
# Using the 'head' method please display the first 5 rows
iris.head(5)
```

```python
# Using “dropna” check if the database contains missing data
iris.dropna().count()
```

```python
# Sort the data in ascending order relative to the second column
iris.sort_values(by=['sepal width (cm)'])
```

```python
# Find the minimum and maximum petal length (column 3 - petal length). Give the index of each of these values.
print('Min value: ' + str(iris.iloc[:, 2].min()))
print('Min value index: ' + str(iris.iloc[:, 2].idxmin()))
print('')
print('Max value: ' + str(iris.iloc[:, 2].max()))
print('Max value index: ' + str(iris.iloc[:, 2].idxmax()))
```

```python
# Calculate the standard deviation for each column
print('Standard Deviation of Column 1 : \t' + str(iris.iloc[:, 0].std()))
print('Standard Deviation of Column 2 : \t' + str(iris.iloc[:, 1].std()))
print('Standard Deviation of Column 3 : \t' + str(iris.iloc[:, 2].std()))
print('Standard Deviation of Column 4 : \t' + str(iris.iloc[:, 3].std()))
print('Standard Deviation of Column 5 : \t' + str(iris.iloc[:, 4].std()))
```

```python
# Identify rows for which the length of the flower cup (sepal length) is greater than the average length (the whole set)

iris[iris.iloc[:, 2] > iris.iloc[:, 2].mean()]
```
